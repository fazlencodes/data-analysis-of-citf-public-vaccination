```python
df_vax_perak = df_vax[df_vax["state"]=="Perak"]
df_vax_perlis = df_vax[df_vax["state"]=="Perlis"]
df_vax_johor = df_vax[df_vax["state"]=="Johor"]
df_vax_kedah = df_vax[df_vax["state"]=="Kedah"]
df_vax_kelantan = df_vax[df_vax["state"]=="Kelantan"]
df_vax_melaka = df_vax[df_vax["state"]=="Melaka"]
df_vax_n9 = df_vax[df_vax["state"]=="Negeri Sembilan"]
df_vax_pahang = df_vax[df_vax["state"]=="Pahang "]
df_vax_pp = df_vax[df_vax["state"]=="Pulau Pinang"]
df_vax_sabah = df_vax[df_vax["state"]=="Sabah"]
df_vax_terengganu = df_vax[df_vax["state"]=="Terengganu"]
df_vax_kl = df_vax[df_vax["state"]=="W.P. Kuala Lumpur"]
df_vax_labuan = df_vax[df_vax["state"]=="W.P. Labuan"]
df_vax_putrajaya = df_vax[df_vax["state"]=="W.P. Putrajaya"]

```

```python
plt.figure(figsize=(15,15))
plt.plot(df_vax_sarawak["total_cumul"],label="Sarawak")
plt.plot(df_vax_selangor["total_cumul"],label="Selangor")
plt.plot(df_vax_perak["total_cumul"],label="Perak")

plt.plot(df_vax_perlis["total_cumul"],label="Perlis")
plt.plot(df_vax_johor["total_cumul"],label="Johor")
plt.plot(df_vax_kedah["total_cumul"],label="Kedah")
plt.plot(df_vax_kelantan["total_cumul"],label="Kelantan")
plt.plot(df_vax_melaka["total_cumul"],label="Melaka")
plt.plot(df_vax_n9["total_cumul"],label="Negeri Sembilan")
plt.plot(df_vax_pahang["total_cumul"],label="Pahang")
plt.plot(df_vax_pp["total_cumul"],label="Pulau Pinang")
plt.plot(df_vax_sabah["total_cumul"],label="Sabah")
plt.plot(df_vax_terengganu["total_cumul"],label="Terengganu")
plt.plot(df_vax_kl["total_cumul"],label="W.P. Kuala Lumpur")
plt.plot(df_vax_labuan["total_cumul"],label="W.P. Labuan")
plt.plot(df_vax_putrajaya["total_cumul"],label="W.P. Putrajaya")
         

plt.xlabel("date")
plt.ylabel("cumulative of total doses")
plt.title("cumulative doses for every states")
plt.legend()
plt.show()
```
